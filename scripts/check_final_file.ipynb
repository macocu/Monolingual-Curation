## Check new format

```python
from prevert import dataset
import pandas as pd

dataset_path = "/data/monolingual/MaCoCu-ca-1.0.xml"
dset=dataset(dataset_path)
```

```python
doc_meta_list = []

for doc in dset:
        doc_meta_list.append(doc.meta) # all attributes

df = pd.DataFrame(doc_meta_list)
df
```

```python
df.describe()
```

```python
# For each attribute, check statistics:
for i in ['id', 'title', 'crawl_date', 'lang_distr', 'url', 'domain', 'file_type', 'lm_score']:
    print(df[i].value_counts()[:10].to_markdown())
```

```python
dset=dataset(dataset_path)

p_meta_list = []

for doc in dset:
        for par in doc:
            p_meta_list.append(par.meta) # all attributes

p_df = pd.DataFrame(p_meta_list)
p_df
```

```python
# For each attribute, check statistics:
for i in ['id', 'quality', 'lm_score', 'sensitive', 'heading', 'lang']:
    print(p_df[i].value_counts()[:10].to_markdown())
```

```python
p_df.describe(include="all")
```
